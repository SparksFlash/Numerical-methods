```python
def simpsons_3_8_rule(fun, a, b):
    a, b = min(a, b), max(a, b)
    h = (b - a) / 3
    return (3 * h / 8) * (fun(a) + 3 * fun(a + h) + 3 * fun(a + 2 * h) + fun(b))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# example function (replace with your function if needed)
def fun(x):
	return np.sin(x)

a, b, c, d = 0, 0.8/3, 0.8/3*2, 0.8
array = np.linspace(a, d, 200)

print("Simpson's 3/8 rule:", simpsons_3_8_rule(fun, a, d))

cs = CubicSpline([a, b, c, d], [fun(a), fun(b), fun(c), fun(d)])
plt.plot(array, fun(array), label="f(x)")
plt.plot(array, cs(array), '--', label="Cubic Spline through nodes")
plt.scatter([a, b, c, d], [fun(a), fun(b), fun(c), fun(d)], color='green')
plt.fill_between(array, cs(array), alpha=0.3)

plt.legend()
plt.show()
```
