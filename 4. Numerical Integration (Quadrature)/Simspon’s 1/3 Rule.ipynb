```python
def simpsons_1_3_rule(fun, a, b):
    return (b - a) / 6 * (fun(a) + 4 * fun((a + b) / 2) + fun(b))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# define the integrand (vectorized)
fun = lambda x: np.sin(x)

# plotting grid
array = np.linspace(0, 0.8, 300)

print(simpsons_1_3_rule(fun, 0, 0.8))

plt.plot(array, fun(array), label="f(x)")
plt.scatter([0, 0.4, 0.8], [fun(0), fun(0.4), fun(0.8)], color='green')
cs = CubicSpline([0, 0.4, 0.8], [fun(0), fun(0.4), fun(0.8)])
plt.plot(array, cs(array), '--', color='red', label='simpsons_1_3_rule')
plt.fill_between(array, cs(array), color='red', alpha=0.3)
plt.legend()
```
