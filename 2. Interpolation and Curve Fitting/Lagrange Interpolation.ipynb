```python
def lagrange(x, y, x_n):
    n = len(x)
    res = 0
    for i in range(n):
        temp = y[i]
        for j in range(n):
            if i != j:
                temp *= x_n - x[j]
                temp /= x[i] - x[j]
        res += temp
    return res

# Example input data (define x, y and the query point x_n)
x = [0.0, 1.0, 2.0]
y = [1.0, 3.0, 2.0]
x_n = 1.5

y_n_lagrange = lagrange(x, y, x_n)
print(f"Lagrange Interpolated value at x = {x_n} is y = {y_n_lagrange}")
```
