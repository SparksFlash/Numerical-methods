```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Generate sample data
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.rand(100, 1)
```

```python
# Gradient Descent with error tracking
def gradient_descent(x, y, m=0, b=0, learning_rate=0.01, epochs=1000):
    n = len(y)
    errors = []   # store error values
    for _ in range(epochs):
        y_pred = m * x + b
        error = np.mean((y - y_pred)**2)   # Mean Squared Error (MSE)
        errors.append(error)

        dm = (-2/n) * np.sum(x * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)
        m -= learning_rate * dm
        b -= learning_rate * db
    return m, b, errors
```

```python
# Run Gradient Descent
m, b, errors = gradient_descent(x, y)
```

```python
# Plot Data and Best Fit Line
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(x, y, label="Data points")
plt.plot(x, m*x + b, color='red', label=f"y={m:.2f}x+{b:.2f}")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gradient Descent: Best Fit Line')
plt.legend()
```

```python
# Plot Error Decreasing
plt.subplot(1,2,2)
plt.plot(errors, color='blue')
plt.xlabel('Epoch')
plt.ylabel('MSE (Error)')
plt.title('Error decreasing during Gradient Descent')
plt.grid()

plt.tight_layout()
plt.show()
```
