```python
# Newton's Backward Interpolation Method
def backward_interpolation(x_values, y_values, x):
    n = len(x_values)
    diff_table = [[0 for _ in range(n)] for _ in range(n)]
    
    # Initialize the first column of the difference table
    for i in range(n):
        diff_table[i][0] = y_values[i]
    
    # Fill the difference table
    for j in range(1, n):
        for i in range(n - 1, j - 1, -1):
            diff_table[i][j] = diff_table[i][j - 1] - diff_table[i - 1][j - 1]
    
    # Calculate the interpolation value
    result = diff_table[n - 1][0]
    h = x_values[1] - x_values[0]
    u = (x - x_values[n - 1]) / h
    
    # Calculate the terms of the interpolation polynomial
    u_term = 1
    for i in range(1, n):
        u_term *= (u + i - 1) / i
        result += u_term * diff_table[n - 1][i]
    
    return result
```
