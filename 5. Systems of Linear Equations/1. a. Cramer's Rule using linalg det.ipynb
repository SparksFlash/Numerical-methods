```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
A0 = np.array([[3,4,-2],
              [-4,1,5],
              [5,-3,-1]])

A1 = np.array([[-9,4,-2],
              [11,1,5],
              [16,-3,-1]])

A2 = np.array([[3,-9,-2],
              [-4,11,5],
              [5,16,-1]])

A3 = np.array([[3,4,-9],
              [-4,1,11],
              [5,-3,16]])
```

```python
A = np.linalg.det(A0)
Ax = np.linalg.det(A1)
Ay = np.linalg.det(A2)
Az = np.linalg.det(A3)
```

```python
x = Ax / A
y = Ay / A
z = Az / A
```

```python
print(x,y,z)
```

```python
df = pd.DataFrame({'x': [x], 'y': [y], 'z': [z]})
```

```python
sns.scatterplot(data=df)
```

```python
sns.barplot(x=df.columns, y=df.iloc[0].values)
plt.ylabel('Value')
plt.title('Values of x, y, z')
plt.show()
```
