```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 3 - 4 * x + 1
```

```python
def bisection_method(f, a, b, tol=1e-5, max_iter=1000):
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) and f(b) must have equal sign")
    
    for _ in range(max_iter):
        mid = (a + b)/2

        if abs(f(mid)) < tol:
            return mid
        
        if f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid

    return (a + b) / 2.0
```

```python
root = bisection_method(f, 1, -1)
# x = np.linspace(root - 2, root + 2, 400)
x = np.arange(-10, 10, 0.1)

plt.plot(x , f(x), label="f(x) = x^3 - 4x + 1")
plt.scatter(root, f(root), color='Blue')
plt.axvline(root,color='Purple',linestyle='--',label=f'x = {root:.5f}')

plt.axhline(0, color='green', linestyle='-')
plt.axvline(0, color='red', linestyle='-')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f"Bisection method root: {root: .5f}")
plt.grid()
plt.legend()
```
