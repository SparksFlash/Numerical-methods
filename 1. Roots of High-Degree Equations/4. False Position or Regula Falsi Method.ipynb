```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 3 - 4 * x + 1
```

```python
def false_position(f, a, b, tol = 1e-5 ,max_iter=1000):
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) and f(b) must have opposite signs.")
    
    for _ in range(max_iter):
        c = a - (f(a) * (b - a)) / (f(b) - f(a))

        if abs(f(c)) < tol:
            return c
        
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    return RuntimeError("Method did not converge within max_iter")
```

```python
root = false_position(f,-1, 1)
x = np.arange(-10, 10, 0.1)

plt.plot(x,f(x),label='f(x) = x^3 - 4x + 1')
plt.scatter(root, f(root),color='blue')
plt.axvline(root, color='purple',linestyle='--', label=f"x = root({root:.5f})")

plt.axhline(0, color='green', linestyle='--')
plt.axvline(0, color='red', linestyle='--')
 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.title(f"False position method root {root: .5f}")
plt.legend()
```
